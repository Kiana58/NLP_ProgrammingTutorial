```python
import math
from collections import defaultdict
from sklearn.externals import joblib
counts = defaultdict(int)
total_count = 0
```

```python
# テスト
# training_file = open("../test/01-train-input.txt","r")
# test_file = open("../test/01-test-input.txt","r")

# 本番
training_file = open("../data/wiki-en-train.word","r")
test_file = open("../data/wiki-en-test.word","r")

training = training_file.readlines()
test = test_file.readlines()
```

```python
for line in training:
    words = line.split()
    words.append("</s>")
    for word in words:
        counts[word] += 1
        total_count += 1
```

```python
probabilities = defaultdict(int)
for word,count in counts.items():
    probability = float(count)/total_count
    probabilities[word] = probability
```

```python
lambda_1 = 0.95
lambda_unknown = 1-lambda_1
V = 1000000; W = 0; H = 0; Unknown = 0

for line in test:
    words = line.split()
    words.append("</s>")
    for word in words:
        W += 1
        P = lambda_unknown/V
        if word in probabilities:
            P += lambda_1*probabilities[word]
        else:
            Unknown += 1
        H += -math.log(P,2)

print "entropy =",H/W
print "coverage =",float(W-Unknown)/W
```
